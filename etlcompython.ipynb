```python
import pandas as pd
import google.generativeai as genai
from google.api_core import retry

# --- CONFIGURAÇÃO ---
GOOGLE_API_KEY = "SUA_KEY" # <--- Cole sua chave aqui novamente

# Configura a conexão (usando REST para evitar travamentos)
genai.configure(api_key=GOOGLE_API_KEY, transport='rest')

# --- A CORREÇÃO ESTÁ AQUI ---
# Estamos usando o nome exato que apareceu na sua lista
model_name = 'models/gemini-2.0-flash'
model = genai.GenerativeModel(model_name)

# 1. DADOS (Simulação do CSV lido)
dados = [
    {'Nome': 'Ana', 'Cargo': 'Desenvolvedora', 'Nota': 9.5},
    {'Nome': 'Carlos', 'Cargo': 'Analista', 'Nota': 7.0},
    {'Nome': 'Beatriz', 'Cargo': 'Designer', 'Nota': 5.5},
    {'Nome': 'João', 'Cargo': 'Gerente', 'Nota': 8.0}
]

# 2. TRANSFORM (Gerar feedback)
def gerar_feedback(funcionario):
    prompt = (
        f"Aja como um especialista em RH. Escreva um feedback curto, profissional e direto (máximo 1 frase) "
        f"para {funcionario['Nome']}, que é {funcionario['Cargo']} e teve nota {funcionario['Nota']}."
    )
    try:
        # Generate content
        response = model.generate_content(prompt)
        return response.text.strip()
    except Exception as e:
        return f"Erro: {str(e)}"

# --- EXECUÇÃO ---
print(f"Iniciando processamento com o modelo {model_name}...")

for func in dados:
    print(f"Processando: {func['Nome']}...")
    func['Feedback_IA'] = gerar_feedback(func)

# 3. RESULTADO
df_final = pd.DataFrame(dados)
pd.set_option('display.max_colwidth', None)

print("\n--- SUCESSO! CONFIRA OS RESULTADOS ABAIXO ---")
display(df_final[['Nome', 'Feedback_IA']])
```
